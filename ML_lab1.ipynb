Objective:
To practice selecting columns and filtering rows using Pandas.

Tasks:
1. Select score-related columns.
2. Filter students scoring above 70 in math.
3. Filter data based on gender.
4. Count number of students in each category.




```python
# 1. Select score-related columns.
score_columns = data[['math_score', 'science_score', 'english_score', 'overall_score']]
print("\nScore-related columns (first 5 rows):\n", score_columns.head())

# 2. Filter students scoring above 70 in math.
students_high_math = data[data['math_score'] > 70]
print("\nStudents with math score > 70 (first 5 rows):\n", students_high_math.head())
print(f"\nNumber of students with math score > 70: {len(students_high_math)}")

# 3. Filter data based on gender (e.g., 'female' students).
female_students = data[data['gender'] == 'female']
print("\nFemale students (first 5 rows):\n", female_students.head())
print(f"\nNumber of female students: {len(female_students)}")

# 4. Count number of students in each category (e.g., gender, school_type).
print("\nStudent count by gender:\n", data['gender'].value_counts())
print("\nStudent count by school type:\n", data['school_type'].value_counts())
```

```python

average_scores = data[['math_score', 'science_score', 'english_score', 'overall_score']].mean()

plt.bar(average_scores.index, average_scores.values)
plt.title('Average Scores Across Subjects')
plt.xlabel('Subjects')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```


Objective:
To analyze student performance based on different groups.

Tasks:
1. Group scores by gender.
2. Group scores by parental education.
3. Visualize group-wise average scores using bar charts.


```python
grouped_by_gender = data.groupby('gender')[['math_score', 'science_score', 'english_score', 'overall_score']].mean()
print( grouped_by_gender)
```

```python
grouped_by_parent_education = data.groupby('parent_education')[['math_score', 'science_score', 'english_score', 'overall_score']].mean()
print("Average scores grouped by parental education:\n", grouped_by_parent_education)
```

```python
# Visualize average scores grouped by gender
grouped_by_gender.plot(kind='bar', figsize=(10, 6))
plt.title('Average Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Visualize average scores grouped by parental education
grouped_by_parent_education.plot(kind='bar', figsize=(12, 7))
plt.title('Average Scores by Parental Education')
plt.xlabel('Parental Education')
plt.ylabel('Average Score')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
